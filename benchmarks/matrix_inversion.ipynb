```python
import numpy as np, matplotlib.pyplot as plt
from timeit import default_timer as timer
from scipy.linalg import cho_solve, cho_factor

def inv_spd(A):
    c, lower = cho_factor(A, lower=True, check_finite=False)
    A_inv = cho_solve((c, lower), np.eye(A.shape[0]), check_finite=False)
    return A_inv
```

```python
size = 5000
mat = np.load(
    "../data/out/K_train_train_128_samples_30000x30000.npy"
    )[:size, :size]
```

```python
n = 13000
np.mean(np.abs(inv_spd(mat[:n, :n])))
```

```python
# Numpy inverse
t0 = timer()
mat_inv_1 = np.linalg.inv(mat)
t1 = timer()
print(f"Time for np inverse for n = {size}:", t1 - t0)

# Scipy symmetric inverse
t0 = timer()
mat_inv_2 = inv_spd(mat)
t1 = timer()
print(f"Time for scipy inverse for n = {size}:", t1 - t0)
```
