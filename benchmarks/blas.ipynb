```python
import numpy as np
from scipy.linalg.blas import dsyrk, dgemm
```

```python
a = np.array(range(5))
b = np.array([10, 20, 30])
c = np.zeros((5, 5))
d = np.zeros((3, 5))
```

```python
dsyrk(a=a, c=c.T, alpha=1.0, beta=1.0, overwrite_c = True)
```

```python
dgemm(a=a, b=b, c=d.T, trans_b=True, alpha=1.0, beta=1.0, overwrite_c=True)
```
